## Cleaning Part 2

```python
import pandas as pd
df_combined = pd.read_csv("main_data1.csv")
```

```python

# Invertir las variables según las instrucciones
def invert_values(series):
    return series.map({1: 4, 2: 3, 3: 2, 4: 1})

# Aplicar la inversión a las columnas indicadas
columns_to_invert = [
    'autoritario_problemas', 'control_medios', 'confianza_ffa', 
    'confianza_polis', 'confianza_iglesia', 'confianza_congreso', 
    'confianza_gob', 'confianza_pj', 'confianza_partidos', 
    'confianza_presi', 'confianza_orga_elector', 'democracia_mejor', 
    'satisfaccion', 'interes',"satisfaccion"
]

for col in columns_to_invert:
    df_combined[col] = invert_values(df_combined[col])

# Invertir los valores de situ_eco (1-5)
df_combined['situ_eco'] = df_combined['situ_eco'].map({1: 5, 2: 4, 3: 3, 4: 2, 5: 1})

# Recodificar izq_der
df_combined['izq_der'] = df_combined['izq_der'].map(
    {0: 1, 1: 1, 2: 1, 3: 1, 4: 2, 5: 2, 6: 2, 7: 3, 8: 3, 9: 3, 10: 3}
)

# Crear variable dummy para izquierda (izq_der == 1)
df_combined['izq_dummy'] = (df_combined['izq_der'] == 1).astype(int)

# Crear variable dummy para derecha (izq_der == 3)
df_combined['der_dummy'] = (df_combined['izq_der'] == 3).astype(int)


# Convertir Sexo a dummy (1: Hombre, 0: Mujer)
df_combined['sexo_dummy_hombre'] = (df_combined['SEXO'] == 1).astype(int)

# Convertir gob_militar a dummy
df_combined['gob_militar_dummy'] = (df_combined['gob_militar'] == 1).astype(int)

df_combined['situ_eco_dummy'] = df_combined['situ_eco'].apply(lambda x: 1 if x in [3, 4] else 0)
df_combined = df_combined.drop('situ_eco', axis=1)
df_combined = df_combined.drop('gob_militar', axis=1)
df_combined = df_combined.drop('SEXO', axis=1)
df_combined = df_combined.drop('izq_der', axis=1)

df_combined['interes_dummy'] = df_combined['interes'].apply(lambda x: 1 if x in [3, 4] else 0)

df_combined = df_combined.drop('interes', axis=1)
df_combined = df_combined.drop('confianza_iglesia', axis=1)
```

```python
# Crear la variable dummy de corrupción
df_combined['corrupcion_dummy'] = 0
df_combined.loc[(df_combined['Año'] == 2020) & (df_combined['problema'] == 34), 'corrupcion_dummy'] = 1
df_combined.loc[(df_combined['Año'] == 2023) & (df_combined['problema'] == 35), 'corrupcion_dummy'] = 1

# Crear la variable dummy de criminalidad
df_combined['criminalidad_dummy'] = 0
df_combined.loc[(df_combined['Año'] == 2020) & (df_combined['problema'].isin([24, 25])), 'criminalidad_dummy'] = 1
df_combined.loc[(df_combined['Año'] == 2023) & (df_combined['problema'].isin([26, 27])), 'criminalidad_dummy'] = 1
df_combined = df_combined.drop('problema', axis=1)
```

```python
import numpy as np
df_combined.dropna()
df_combined = df_combined.dropna(how='any')  # Elimina las filas con NAs
df_combined = df_combined.replace("", np.nan).dropna(how='any')  # Elimina las filas con valores vacíos
```
